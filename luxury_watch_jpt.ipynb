```python
import pandas as pd

df = pd.read_csv("luxury_watch.csv", sep=',', index_col=0)
df
```

```python
df['Case_Diameter'] = df['Case_Diameter'].astype(float)
df['Case_Thickness'] = df['Case_Thickness'].astype(float)
df['Band_Width'] = df['Band_Width'].astype(float)
df['Price'] = df['Price'].str.replace(',', '').astype(float)
df['Power_Reserve'] = df['Power_Reserve'].str.extract('(\d+)').astype(float)
df
```

```python
# Statistiques descriptives
df.describe()
```

```python
# Analyse des valeurs manquantes
df.isnull().sum()
```

```python
print(df.columns)
```

```python
df = df.reset_index(drop=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution des prix
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True)
plt.title('Distribution des prix')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Distribution des diamÃ¨tres de boÃ®tier
plt.figure(figsize=(10, 6))
sns.histplot(df['Case_Diameter'], kde=True)
plt.title('Distribution des diamÃ¨tres de boÃ®tier')
plt.xlabel('Case Diameter (mm)')
plt.ylabel('Frequency')
plt.show()

# Boxplot des prix par marque
plt.figure(figsize=(12, 8))
sns.boxplot(x='Brand', y='Price', data=df)
plt.title('Prix par marque')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()

# Pairplot pour voir les relations entre les caractÃ©ristiques numÃ©riques
sns.pairplot(df[['Case_Diameter', 'Case_Thickness', 'Band_Width', 'Power_Reserve', 'Price']])
plt.show()
```

```python
# CorrÃ©lation entre les variables numÃ©riques
correlation_matrix = df[['Case_Diameter', 'Case_Thickness', 'Band_Width', 'Power_Reserve', 'Price']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de corrÃ©lation')
plt.show()

# Scatter plot du prix en fonction du diamÃ¨tre de boÃ®tier
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Case_Diameter', y='Price', data=df, hue='Brand')
plt.title('Prix en fonction du diamÃ¨tre de boÃ®tier')
plt.xlabel('Case Diameter (mm)')
plt.ylabel('Price')
plt.show()
```
